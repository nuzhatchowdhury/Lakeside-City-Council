```python
import pandas as pd

# Load files
registered_voters = pd.read_csv("registered_voters.csv")
northgate = pd.read_csv("northgate_early_inperson.csv")
mail = pd.read_excel("mail_ballots.xlsx")

# Convert IDs to strings
registered_voters["SOS_VoterID"] = registered_voters["SOS_VoterID"].astype(str)
northgate["CertificateNumber"] = northgate["CertificateNumber"].astype(str)
mail["VUID"] = mail["VUID"].astype(str)

runoff_ids = set(northgate["CertificateNumber"]) | set(mail["VUID"])

# Find voters who participated
runoff_voters = registered_voters[
    registered_voters["SOS_VoterID"].isin(runoff_ids)
]

print(runoff_voters.head())
print("Runoff voters:", len(runoff_voters))
```

```python
import pandas as pd 
registered_voters = pd.read_csv("registered_voters.csv")
registered_voters["SOS_VoterID"] = registered_voters["SOS_VoterID"].astype(str)

runoff_voters = registered_voters[
    registered_voters["SOS_VoterID"].isin(runoff_ids)
]
```

```python
runoff_count = (
    runoff_voters
    .groupby("precinct_raw")
    .size()
    .reset_index(name="runoff_voters")
)
```

```python
turnout = registered_count.merge(
    runoff_count,
    on="precinct_raw",
    how="left"
)

turnout["runoff_voters"] = turnout["runoff_voters"].fillna(0)
```

```python
turnout["turnout_pct"] = (
    turnout["runoff_voters"] /
    turnout["registered_voters"] * 100
).round(2)
```
